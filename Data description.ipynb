# Description du jeu de données

## Soil Wetness Index (SWI)

Le **SWI (Soil Wetness Index)** est un indice modélisé par Météo-France qui représente l’état de l’humidité des sols jusqu’à environ 2 mètres de profondeur et sert à évaluer les sécheresses dans le cadre des catastrophes naturelles. Parmi les facteurs variant dans le temps, la température de l'air et l'évapotranspiration sont des régulateurs importants de la dynamique de l'humidité du sol. Il est calculé à l’échelle de la France métropolitaine sur une grille régulière d’environ 8 km × 8 km, issue du modèle hydrométéorologique SIM, permettant de représenter spatialement les variations de l’humidité des sols sur l’ensemble du territoire.

Il s’agit d’un indice sans unité représentant l'état de l'humidité du sol par rapport à ses bornes hydriques, défini par :

`SWI = (W - W_wilt) / (W_fc - W_wilt)`

où `W` est le contenu en eau du sol, `W_wilt` le point de flétrissement et `W_fc` la capacité au champ. Il varie généralement entre 0 (sol très sec) et 1 (sol à capacité au champ), avec des valeurs possibles en dehors de cet intervalle en conditions extrêmes.

Le **SWI uniforme** est une version simplifiée de cet indice où les caractéristiques du sol et de la végétation sont identiques partout, afin de comparer les territoires uniquement selon les conditions météorologiques.

## Données

Notre jeu de données couvre la période **1960–2024** et contient :
- l’indice d’humidité des sols **SWI uniforme**,
- des variables météorologiques : précipitations (`PRENEI`, `PRELIQ`), température (`T`), vent (`FF`),
- des flux d’eau : `EVAP`, `ETP`,
- des données liées aux sols argileux,
- des variables hydrologiques : ruissellement, drainage, neige.
