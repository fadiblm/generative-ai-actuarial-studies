# Climate Risk Indicators Notebook
## Insurance-oriented climate indicators and where to download the data

This notebook summarizes a practical list of climate indicators that are highly relevant for climate-risk insurance work, especially for hazard modeling, pricing, stress testing, and generative simulation.

For each indicator, you will find:
- a short definition,
- why it matters for insurance,
- the key input variables,
- practical download hints.

---

## 1. Wind Speed Index

**Definition**  
A wind hazard indicator built from sustained wind speed and/or maximum wind gusts.

**Why it matters for insurance**
- Windstorm losses for residential and commercial property
- Infrastructure damage
- Business interruption

**Typical variables**
- 10 m wind speed
- 10 m u/v wind components
- maximum wind gust (if available)

**Download hints**
- **ERA5 / ERA5-Land (Copernicus CDS)**: global reanalysis for 10 m wind components and many related variables  
- **Météo-France open data**: station observations and climatological wind products for France  
- **E-OBS / national station networks** for Europe-wide observed products

**Notes**
- For pricing work, insurers often derive event-based wind metrics such as maximum gust per event, annual maximum gust, or threshold exceedance counts.

## 2. Extreme Rainfall Index

**Definition**  
A general family of precipitation hazard indicators focused on short-duration or multi-day extremes.

**Why it matters for insurance**
- River flooding
- Flash flooding
- Urban runoff
- Agricultural losses

**Typical variables**
- Total precipitation
- Event duration
- Antecedent wetness (optional)

**Download hints**
- **ERA5 / ERA5 daily statistics** for precipitation time series  
- **Météo-France** for station and gridded rainfall products in France  
- **E-OBS** for European daily precipitation grids

**Notes**
- In pricing and cat modeling, this is often transformed into return levels, exceedance counts, or event severity scores.

## 3. RX1day

**Definition**  
Maximum precipitation accumulated over **1 day** during a given period.

**Why it matters for insurance**
- Key driver of flood and runoff losses
- Common extreme-climate indicator in cat and climate-risk studies

**Formula**
```text
RX1day = max(daily precipitation)
```

**Download hints**
- Compute directly from daily precipitation from **ERA5 daily statistics**, **E-OBS**, or **Météo-France** datasets
- Can also be derived from station observations

**Notes**
- Useful targets: annual RX1day, seasonal RX1day, or location-wise return levels.

## 4. RX5day

**Definition**  
Maximum precipitation accumulated over **5 consecutive days** during a given period.

**Why it matters for insurance**
- Better captures prolonged wet spells than RX1day
- Strongly linked to river flooding and soil saturation

**Formula**
```text
RX5day = max(rolling 5-day precipitation sum)
```

**Download hints**
- Compute from daily precipitation using **ERA5 daily statistics**, **E-OBS**, or **Météo-France**
- Particularly useful for flood-prone regions

**Notes**
- Often paired with soil moisture indicators to model compound flood risk.

## 5. Soil Wetness Index (SWI)

**Definition**  
An index describing the relative wetness of the soil, often used as a proxy for available soil water.

**Why it matters for insurance**
- Drought losses
- Agricultural stress
- Wildfire susceptibility
- Building damage from shrink-swell clay (important in France)

**Typical variables**
- Soil moisture
- Field capacity / climatology
- Root-zone depth (depending on implementation)

**Download hints**
- **Météo-France** publishes monthly soil-wetness data for the CatNat system in France
- **ERA5-Land** provides land variables including soil moisture
- **European Drought Observatory (EDO/GDO)** provides soil-moisture anomaly layers and drought indicators

**Notes**
- If a direct SWI product is unavailable, many teams compute soil-moisture anomalies or percentile-based wetness indices.

## 6. Forest Drought Index (IS)

**Definition**  
An indicator of forest dryness used to characterize how dry the forest environment is.

**Why it matters for insurance**
- Wildfire risk
- Forestry risk
- Climate stress on ecosystems

**Typical variables**
- Temperature
- Rainfall
- Soil/vegetation dryness
- Fuel moisture proxies

**Download hints**
- Often available through national fire-danger systems or derived from meteorological reanalyses
- In Europe, combine **ERA5 / ERA5-Land** with fire-danger products from **EFFIS**
- For France, use Météo-France and forest-fire operational products when available

**Notes**
- Exact operational formulations differ across institutions.

## 7. Fire Weather Index (FWI / IFM)

**Definition**  
A standard meteorological fire-danger indicator that summarizes how favorable the atmosphere is to wildfire ignition and spread.

**Why it matters for insurance**
- Directly relevant to wildfire hazard
- Widely used in operational fire-risk monitoring

**Typical variables**
- Temperature
- Relative humidity
- Wind speed
- Precipitation

**Download hints**
- **EFFIS (Copernicus)** provides FWI and related indices such as ISI, BUI, FFMC, DMC, and DC
- Can also be recomputed from reanalysis weather variables

**Notes**
- FWI is one of the most directly useful operational climate indicators for wildfire insurance.

## 8. Haines Index

**Definition**  
An atmospheric stability and dryness index designed to indicate the potential for large or erratic wildfire growth.

**Why it matters for insurance**
- Complements FWI by capturing atmospheric conditions conducive to plume-driven fire behavior

**Typical variables**
- Temperature at selected pressure levels
- Moisture / dew point depression at selected pressure levels

**Download hints**
- Can be computed from **ERA5 pressure-level data**
- Also available through some meteorological and fire-weather systems

**Notes**
- Especially useful in wildfire-risk analytics when combined with FWI.

## 9. Temperature

**Definition**  
A core climate variable used both directly and as input to many hazard indices.

**Why it matters for insurance**
- Heatwaves
- Crop stress
- Mortality / health claims
- Energy demand stress
- Input to drought and fire metrics

**Typical variables**
- 2 m air temperature
- Daily minimum / maximum temperature
- Temperature anomalies

**Download hints**
- **ERA5 / ERA5-Land**
- **Météo-France**
- **E-OBS** for Europe
- **CMIP6** for climate projections

**Notes**
- Temperature is one of the easiest and most robust variables to access and derive indices from.

## 10. Convective Storm Severity Index

**Definition**  
A severe-convection indicator used for hail, thunderstorms, and tornado-like environments.

**Why it matters for insurance**
- Hail losses
- Severe thunderstorm losses
- Tornadic environments
- Localized but very costly events

**Typical variables**
- **CAPE** (Convective Available Potential Energy)
- **Wind shear**
- Lifted Index / CIN / storm-relative helicity (optional)

**Download hints**
- **ERA5** provides CAPE and wind fields needed to build severe-convection proxies
- Operational severe-weather products may exist from national meteorological agencies
- For Europe, combine reanalysis with lightning / hail / radar datasets where available

**Notes**
- There is no single universal index: insurers often build their own severe-storm score from CAPE, shear, and related variables.

## Suggested official / practical download portals

### Copernicus Climate Data Store (CDS)

- ERA5 daily statistics
- CMIP6 climate projections

### Météo-France open data
- Station observations
- Climatological products
- Soil wetness data for France (CatNat-oriented products)

### Copernicus Emergency Management Service
- **EFFIS** for wildfire indicators such as FWI
- **European Drought Observatory (EDO/GDO)** for soil-moisture anomalies and drought indicators

### Other common sources
- **E-OBS** for European gridded daily observations
- National radar / lightning / hail archives
- River discharge portals for flood-oriented extensions

## Practical recommendation for generative modeling

In many insurance workflows, it is better to generate **fundamental weather variables** first:
- temperature,
- precipitation,
- wind,
- humidity,
- soil moisture,

and then derive the indices deterministically.

Why?
1. It preserves more physical consistency.
2. It lets you recompute several indices from the same simulation.
3. It is closer to how catastrophe and climate-risk models are built in practice.
