<h4>3.1 Box–Cox Transformation</h4>

<p>
The <strong>Box–Cox transformation</strong> is a parametric power transformation used to make data more 
<strong>Gaussian</strong>, stabilize <strong>variance</strong>, and improve the performance of statistical 
or machine learning models that assume normality.
</p>

<hr>

<h5>Definition</h5>

<p>
For a strictly positive variable \(X > 0\), the Box–Cox transformation is:
</p>

<p style="text-align:center;">
\[
X^{(\lambda)} =
\begin{cases}
\frac{X^\lambda -1}{\lambda} & \text{if } \lambda \neq 0 \\
\log(X) & \text{if } \lambda = 0
\end{cases}
\]
</p>

<p>where:</p>

<ul>
<li>\(\lambda\) is a parameter estimated from the data</li>
<li>it controls the shape of the transformation</li>
</ul>

<hr>

<h4>Intuition</h4>

<p>
It generalizes many common transformations:
</p>

<table>
<tr>
<th>&lambda;</th>
<th>Transformation</th>
</tr>

<tr>
<td>1</td>
<td>identity</td>
</tr>

<tr>
<td>0.5</td>
<td>square root</td>
</tr>

<tr>
<td>0</td>
<td>log</td>
</tr>

<tr>
<td>-1</td>
<td>inverse</td>
</tr>

<tr>
<td>2</td>
<td>square</td>
</tr>

</table>

<p>
Thus, Box–Cox automatically selects the most appropriate power transformation.
</p>

<hr>

<h5>Why use it?</h5>

<p>It is useful when data are:</p>

<ul>
<li>skewed</li>
<li>heteroscedastic</li>
<li>heavy-tailed</li>
<li>non-Gaussian</li>
</ul>

<p>It can improve:</p>

<ul>
<li>regression quality</li>
<li>generative modeling</li>
<li>statistical inference</li>
</ul>


<h5>How &lambda; is chosen</h5>

<p>&lambda; is estimated by maximum likelihood assuming transformed data should be Gaussian.</p>

`scipy` does this automatically.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

X = np.random.exponential(size=1000)

X_transformed, lambda_ = stats.boxcox(X)

print("Lambda:", lambda_)

plt.subplot(121)
plt.hist(X)
plt.title("Original")

plt.subplot(122)
plt.hist(X_transformed)
plt.title("Box-Cox")

plt.show()
```

<h5>Important limitations</h5>

1. _Only works for positive data._ If data contains zeros or negatives, one should ass a constant

`X = X + abs(X.min()) + 1e-6`

2. Alternative: _Yeo–Johnson transform_

`from sklearn.preprocessing import PowerTransformer`

`pt = PowerTransformer(method='yeo-johnson')`

`X_trans = pt.fit_transform(X.reshape(-1,1))`
